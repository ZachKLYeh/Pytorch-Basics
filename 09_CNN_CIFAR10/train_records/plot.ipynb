```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import re

skiprows = np.array([0, 51, 52, 53, 54, 55, 56])
dfcnn = pd.read_csv('VGGNet.txt',skiprows = skiprows)
dfvgg16 = pd.read_csv('VGG16.txt',skiprows = skiprows)
dfvgg16bn = pd.read_csv('VGG16BN.txt',skiprows = skiprows)
dfvggnet = pd.read_csv('VGGNet.txt',skiprows = skiprows)
dfcnn.columns = ["epoch", "step", "accuracy"]
dfvgg16.columns = ["epoch", "step", "accuracy"]
dfvgg16bn.columns = ["epoch", "step", "accuracy"]
dfvggnet.columns = ["epoch", "step", "accuracy"]

x = np.zeros(50)
y = np.zeros(50)
z = np.zeros(50)
w = np.zeros(50)
epochmatrix = np.arange(1, 51, 1)

for i in range(0,49,1):
    a = re.findall("\d+\.\d+",dfcnn['accuracy'][i])
    b = float(a[0])
    x[i]= b
    
for i in range(0,49,1):
    a = re.findall("\d+\.\d+",dfvgg16['accuracy'][i])
    b = float(a[0])
    y[i]= b

for i in range(0,49,1):
    a = re.findall("\d+\.\d+",dfvgg16bn['accuracy'][i])
    b = float(a[0])
    z[i] = b

for i in range(0,49,1):
    a = re.findall("\d+\.\d+",dfvggnet['accuracy'][i])
    b = float(a[0])
    w[i] = b


#plt.plot(epochmatrix, x, label='cnn_loss/step')
plt.plot(epochmatrix, y, label='vgg_loss/step')
plt.plot(epochmatrix, z, label='vggbn_loss/step')
plt.plot(epochmatrix, w, label='torchvgg_loss/step')
plt.xlabel('step')
plt.ylabel('loss')
plt.xscale('linear')
plt.yscale('linear')
plt.title('Plot of step/accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
